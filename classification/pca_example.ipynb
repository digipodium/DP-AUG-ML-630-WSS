```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import  make_pipeline

import plotly.express as px
```

```python
url = 'https://raw.githubusercontent.com/digipodium/Datasets/main/classfication/diabetes.csv'
df = pd.read_csv(url)
df.head()
```

```python
X = df.drop(columns='Outcome')
y = df.Outcome
```

```python
features = PCA(n_components=3).fit_transform(X,y)
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(features, y.values, test_size=.2, random_state=2)
clf = make_pipeline(StandardScaler(), KNeighborsClassifier())
clf.fit(Xtrain, ytrain)
ypred= clf.predict(Xtest)
print('MATRIX\n', confusion_matrix(ytest, ypred))
print('REPORT\n', classification_report(ytest, ypred))
```
