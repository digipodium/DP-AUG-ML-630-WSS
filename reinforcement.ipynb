```python
import pandas as pd
import thompson as rf
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/digipodium/Datasets/main/reinforcement/Ads_CTR_Optimisation.csv')
```

```python
# with the algorithm
result = rf.UCB_random(df)
rf.plot(result)
```

```python
# with ucb algorithm
result = rf.UCB(df)
rf.plot(result)
```

```python
# with Thompson sampling
result = rf.thompson(df)
rf.plot(result)
```
